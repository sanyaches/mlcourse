```python
d = {'No' : False, 'Yes' : True}
dataFrame['International plan'] = dataFrame['International plan'].map(d)
dataFrame.head()
```

```python
dataFrame = dataFrame.replace({'Voice mail plan': d})
dataFrame.head()
```

```python
columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
dataFrame.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

```python
dataFrame.groupby(['Churn', 'International plan'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
pd.crosstab(dataFrame.Churn, dataFrame['International plan'])
```

```python
pd.crosstab(dataFrame.Churn, dataFrame['Voice mail plan'], normalize=True)
```

```python
dataFrame.pivot_table(['Total day calls', 'Total eve calls', 
                       'Total night calls'], 
                      ['Area code'], 
                      aggfunc='mean').head(10)
```

```python
total_calls = dataFrame['Total day calls'] + dataFrame['Total eve calls'] + \
dataFrame['Total night calls'] + dataFrame['Total intl calls']
dataFrame.insert(loc=len(dataFrame.columns), column='Total calls', value=total_calls)
dataFrame.head()
```

```python
dataFrame['Total charge'] = dataFrame['Total day charge'] + dataFrame['Total eve charge'] + \
                            dataFrame['Total night charge'] + dataFrame['Total intl charge']
dataFrame.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
pd.crosstab(dataFrame.Churn, dataFrame['International plan'], margins=True)
```

```python
sns.countplot(x='International plan', hue='Churn', data=dataFrame)
plt.savefig('int_plan_and_churn.png', dpi=300)
```

```python
pd.crosstab(dataFrame.Churn, dataFrame['Customer service calls'], margins=True)
```

```python
sns.countplot(x='Customer service calls', hue='Churn', data=dataFrame)
plt.savefig('serv_calls_and_churn.png', dpi=300)
```

```python
dataFrame['Many_service_calls'] = (dataFrame['Customer service calls'] > 3).astype('int')
pd.crosstab(dataFrame['Many_service_calls'], dataFrame.Churn, margins=True)
```

```python
sns.countplot(x='Many_service_calls', hue='Churn', data=dataFrame)
plt.savefig('many_serv_calls_and_churn.png', dpi=300)
```
