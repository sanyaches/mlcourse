**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

**11. Какой процент мужчин (от всех мужчин) и женщин (от всех женщин), работающих в государственных компаниях, получает >50K? **

**13. Среди всех высокооплачиваемых людей (>50K) больше людей с высшим образованием, которые работают больше 50 часов в неделю или тех, кто не имеет высшего образование и работает не больше 50 часов в неделю? **

```python
richs = data[data.salary == '>50K']
high_educations = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

print ('Высокооплачиваемых людей с высшим образованием, которые работают больше 50 часов в неделю:', len(richs[(richs.education.isin(high_educations)) & (richs['hours-per-week'] > 50)]))
print('Выосокооплачиваемых лентяев: ', len(richs[~(richs.education.isin(high_educations)) & (richs['hours-per-week'] <= 50)]))
```
